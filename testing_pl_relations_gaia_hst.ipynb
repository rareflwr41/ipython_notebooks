```python
import matplotlib.pyplot as mp
import numpy as np
import seaborn as sns
import itertools
```

```python
p1 = np.arange(-1,0.1,0.1)

sns.set(style="darkgrid")
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.0})
sns.set_palette("Set2", 2)

palette = itertools.cycle(sns.color_palette())

slope = -2.342
hst_zp = -1.155
gaia_zp = -0.970

col = next(palette)


mp.plot(p1, slope*(p1)+ hst_zp, ls='--', label="HST zp", color=col)
col = next(palette)
mp.plot(p1, slope*(p1)+ gaia_zp, ls='--', label="Gaia zp", color=col)

mp.xlim(-0.6, -0.1)
mp.ylim(2.0, -2.0)
mp.xlabel('log P (days)')
mp.ylabel('Absolute Magnitude')
mp.suptitle('Comparison of Gaia and HST zero points')
mp.legend(loc='center right', bbox_to_anchor=(1.4, 0.5))

```
